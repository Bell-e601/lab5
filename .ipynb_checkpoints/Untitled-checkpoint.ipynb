Пустое значение в колонке Type 2

```python
pokemons_without_NaN = pokemons.dropna(subset='Type 2')
pokemons_without_NaN
```

Строка с пустым значением удалилась


```python
pokemons.fillna(pokemons_without_NaN['Type 2'].mode().values[0])
```

Пустое значение заменилось на самое распространённое в колонке

-  Создание нового поля вычисленного на основе значений других полей:
- - Через Dataframe.apply

```python
pokemons['Attack Sum'] = pokemons.apply(lambda row: row['Attack'] + row['Sp. Atk'], axis = 1)
pokemons
```

Сложение значений атаки и спец. атаки в новый столбец суммы атаки
